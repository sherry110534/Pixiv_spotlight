```python
#get user id
for k in range(rangenum):
    json_result = aapi.user_following(7611911, restrict='public', offset=k*30)
    for i in range(len(json_result.user_previews)):
        data = {
            'id': "",
            'image': ""
        }
        data['id'] = json_result.user_previews[i].user['id']
        data['image'] = json_result.user_previews[i].user['profile_image_urls']['medium']
        userdata.append(data)
print(len(userdata))
```

```python
#get user id v2
for k in range(rangenum):
    json_result = aapi.user_following(7611911, restrict='public', offset=k*30)
    for i in range(len(json_result.user_previews)):
        data = {
            'id': "",
            'image': ""
        }
        if(json_result.user_previews[i].user['id'] in data_array):
            data['id'] = json_result.user_previews[i].user['id']
            data['image'] = json_result.user_previews[i].user['profile_image_urls']['medium']
            userdata.append(data)
print('total:', len(userdata))
```

```python
# write user data in csv
with open('user_details.csv', 'w', newline='', encoding='utf8') as csvfile:
    field_names = ['user_id','name','profile_image_urls','works','favorites','twitter','job','introduction','homepage']
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    #write user data
    num = 0
    for data in userdata:  
        json_result = api.users(data['id'])
        write_data = {
            'user_id': data['id'],
            'name':json_result.response[0].name,
            'profile_image_urls': data['image'],
            'works': json_result.response[0].stats['works'],
            'favorites': json_result.response[0].stats['favorites'],
            'twitter': json_result.response[0].profile['contacts']['twitter'],
            'job': json_result.response[0].profile['job'],
            'introduction': [json_result.response[0].profile['introduction']],
            'homepage': json_result.response[0].profile['homepage']
        }
        writer.writerow(write_data)
        num = num + 1
    print('complete')
    print('total:', num)
```

```python
#write illusts in csv
with open('illusts_details.csv', 'w', newline='', encoding='utf8') as csvfile:
    field_names = ['illust_id','title','user_id','image_urls','caption','tags','create_date','width','height','total_bookmarks','total_comments']
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    data_num = 0
    total_illusts_num = 0
    for data in userdata:  
        illust_num = 0
        json_result = aapi.user_illusts(data['id'])
        while True: 
            for illust in json_result.illusts:
                total_illusts_num += 1
                write_data = {
                    'illust_id': illust['id'],
                    'title': illust['title'],
                    'user_id': illust['user']['id'],
                    'image_urls': illust['image_urls']['large'],
                    'caption': illust['caption'],
                    'tags': illust['tags'],
                    'create_date': illust['create_date'],
                    'width': illust['width'],
                    'height': illust['height'],
                    'total_bookmarks': illust['total_bookmarks'],
                    'total_comments': illust['total_comments'] 
                }
                writer.writerow(write_data)
                illust_num = illust_num + 1;
            #whether there is next page ot not
            if json_result.next_url == None:
                break
            else:
                next_qs = aapi.parse_qs(json_result.next_url)
                json_result = aapi.user_illusts(**next_qs)
        print('illusts of user:', illust['user']['id'])
        print('total:', illust_num)
        data_num = data_num + 1
    print('complete')
    print('total:', data_num)
    print('total:', total_illusts_num)
    
```

```python
#download square_medium
path =  os.path.join(os.getcwd(), 'square_medium')
size = 'square_medium'
try:
    if not os.path.exists(path):
        os.makedirs(path)
    total_num = 0
    for data in userdata:
        json_result = aapi.user_illusts(data['id'])
        user_illusts_num = 0
        while True:
            #check the number of picture  
            for illust in json_result.illusts:
                user_illusts_num += 1
                total_num += 1
                if(total_num % 3 == 0):
                    time.sleep(2)
                aapi.download(illust.image_urls[size], path = path, name = str(illust.id) + '.jpg')
                print('download ', illust.id, ' complete')
            #whether there is next page ot not
            if json_result.next_url == None:
                break
            else:
                next_qs = aapi.parse_qs(json_result.next_url)
                json_result = aapi.user_illusts(**next_qs)
        print('total number of ', data['id'],' illusts: ', user_illusts_num)
    print('complete')
    print('total: ',total_num)
except Exception as e:
    print(e)
    print('total: ',total_num)
```

```python
#create xls
from openpyxl import Workbook
try:
    field_names = ['illust_id','title','user_id','image_urls','caption','tags','create_date','width','height','total_bookmarks','total_comments']
    total_num = 0
    for data in userdata:
        wb = Workbook()
        ws = wb.active
        ws.append(field_names)
        wb.save(str(data['id']) + '.xlsx') 
        total_num+=1
    print('complete')
    print('total: ',total_num)
except Exception as e:
    print(e)
    print('total: ',total_num)
```
